## Analyzing Data (Before Processing)

```python
preferred_categories = [
    "rec.sport.hockey",
    "sci.med"
]
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
for each_preferred_category in preferred_categories:
    wc = WordCloud().generate(' '.join(full_df[full_df['categories'] == each_preferred_category]['text']))
    plt.figure(figsize=(10, 10))
    plt.title(each_preferred_category, fontsize=30)
    plt.imshow(wc)
```
